# Getting started with matplotlib: Visualizing hospitality data

### Line Graph

```python
import matplotlib.pyplot as plt

# Sample Data: Monthly hotel room bookings
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
bookings = [120, 150, 180, 200, 170, 190]

# Create a simple line chart
plt.plot(months, bookings, color='blue', marker='o')
plt.title("Monthly Hotel Room Bookings")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.show()
```

### Bar Graph

```python
plt.bar(months, bookings, color='skyblue')
plt.title("Monthly Hotel Room Bookings")
plt.xlabel("Month")
plt.ylabel("Bookings")
plt.show()
```

#### When to use:
- Bar charts are great for comparing categories (e.g., bookings per month, revenue per city).

### Pie Chart 

```python
# Market share of booking sources
sources = ['Website', 'Travel Agent', 'Walk-in']
percentages = [50, 35, 15]

plt.pie(percentages, labels=sources, autopct='%1.1f%%', startangle=90)
plt.title("Booking Source Distribution")
plt.show()
```

#### When to use:
- Pie charts show proportions of a whole â€” e.g., how customers book their rooms.

### Scatter Plot



```python
# Relationship between room price and customer rating
price = [100, 120, 150, 180, 200, 220]
rating = [3.5, 4.0, 4.2, 4.3, 4.7, 4.8]

plt.scatter(price, rating, color='orange')
plt.title("Room Price vs Customer Rating")
plt.xlabel("Room Price ($)")
plt.ylabel("Rating (out of 5)")
plt.show()
```

#### When to use:
- Shows relationship between two numeric variables (e.g., higher prices â†’ better ratings).

### Customizing Matplotlib Plots

- **figsize â†’** change plot size.

- **linestyle, color, linewidth â†’** customize appearance.

- **grid(True) â†’** add grid lines for clarity.

```python
plt.figure(figsize=(8,5))
plt.plot(months, bookings, color='green', linestyle='--', marker='o', linewidth=2)
plt.title("Customized Line Plot: Hotel Bookings", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Bookings")
plt.grid(True)
plt.show()
```

# KPIs (Key Performance Indicators) in hospitality domain 

#### **It helps us track how well a hotel or restaurant is doing. Common KPIs include:**

**Occupancy Rate:** % of rooms booked out of total available rooms.

**Average Daily Rate (ADR):** Average revenue earned per occupied room.

**Revenue per Available Room (RevPAR):** Total room revenue divided by total available rooms.

**Customer Satisfaction Score:** Rating from customer reviews.

**Average Length of Stay:** Average number of nights guests stay.

#### Simulated hotel booking data for analysis and visualization

```python
import pandas as pd

# Sample hotel booking data
data = {
    'Date': pd.date_range(start='2023-10-01', periods=10, freq='D'),
    'Rooms_Available': [100]*10,
    'Rooms_Booked': [80, 85, 90, 95, 75, 70, 60, 65, 80, 90],
    'Room_Revenue': [8000, 8500, 9000, 9500, 7500, 7000, 6000, 6500, 8000, 9000],
    'Customer_Rating': [4.5, 4.6, 4.7, 4.4, 4.3, 4.2, 4.1, 4.0, 4.5, 4.6],
    'Length_of_Stay': [2, 3, 2, 1, 4, 3, 2, 2, 3, 1]
}

df = pd.DataFrame(data)
print(df)
```

#### Calculating Occupancy Rate and Visualizing It

Occupancy Rate formula:

 **OccupancyÂ Rate = (RoomsÂ Booked/RoomsÂ Available)Ã—100**


```python
import matplotlib.pyplot as plt

# Calculate Occupancy Rate
df['Occupancy_Rate'] = (df['Rooms_Booked'] / df['Rooms_Available']) * 100

# Plot Occupancy Rate over time
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Occupancy_Rate'], marker='o', linestyle='-')
plt.title('Hotel Occupancy Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Occupancy Rate (%)')
plt.grid(True)
plt.show()
```

**Explanation:**

- This line chart shows how the occupancy rate changes each day.

- Higher occupancy means more rooms booked relative to total rooms, so better utilization.

- A lower occupancy rate might signal lower demand or operational issues.

#### Calculating and Visualizing Average Daily Rate (ADR)

**What is ADR?**

- It tells you the average revenue you earn from each occupied room.
- Itâ€™s a key metric to understand pricing and revenue efficiency.

Formula:

**ð´ð·ð‘… = RoomÂ Revenue / RoomsÂ Booked**


```python
import numpy as np

# Adding some randomness to Room_Revenue to break the pattern
np.random.seed(42)  # for reproducibility
df['Room_Revenue'] = df['Room_Revenue'] * np.random.uniform(0.9, 1.1, size=len(df))

# Recalculate ADR with new Room_Revenue values
df['ADR'] = df['Room_Revenue'] / df['Rooms_Booked']

# Plot ADR with scatter + line
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['ADR'], linestyle='-', color='orange', label='ADR Trend')
plt.scatter(df['Date'], df['ADR'], color='red', label='ADR Points')
plt.title('Average Daily Rate (ADR) Over Time (With Variability)')
plt.xlabel('Date')
plt.ylabel('ADR ($)')
plt.grid(True)
plt.legend()
plt.show()
```

#### Calculating and Visualizing Revenue per Available Room (RevPAR)

**What is RevPAR?**

- It combines occupancy and ADR to show overall room revenue performance.
- It helps understand if youâ€™re maximizing revenue from your total inventory (not just occupied rooms).

Formula:

**ð‘…ð‘’ð‘£ð‘ƒð´ð‘… = (RoomÂ Revenue/ RoomsÂ Available)**

or

**ð‘…ð‘’ð‘£ð‘ƒð´ð‘… = OccupancyÂ Rate Ã— ADR**

```python
# Calculate RevPAR
df['RevPAR'] = df['Room_Revenue'] / df['Rooms_Available']

# Plot RevPAR over time
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['RevPAR'], marker='^', color='green', linestyle='-')
plt.title('Revenue per Available Room (RevPAR) Over Time')
plt.xlabel('Date')
plt.ylabel('RevPAR ($)')
plt.grid(True)
plt.show()
```

##### What does RevPAR show?

- It reflects both how full your hotel is (occupancy) and how well you price rooms (ADR).
- If RevPAR increases, your hotel is generating more revenue per room available.
